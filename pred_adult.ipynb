```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

#Ładowanie pliku csv do ramki danych
df = pd.read_csv("../sztuczna_inteligencja/adult.csv/adult.csv", delimiter=",")
data = [df]
print (df.head())
```

```python
#Szukanie korelacji między kolumnami
def plot_correlation(df, size=15):

    numeric_df = df.select_dtypes(include=['float64', 'int32', 'int64'])
    
    if not numeric_df.empty:
        corr = numeric_df.corr()
        fig, ax = plt.subplots(figsize=(size, size))
        ax.matshow(corr)
        plt.xticks(range(len(corr.columns)), corr.columns)
        plt.yticks(range(len(corr.columns)), corr.columns)
        plt.show()
    else:
        print("No numeric columns in the DataFrame.")
```

```python
plot_correlation(df)
```

ZASTOSOWANIE MODELU UCZENIA SIĘ

Dzielimy dane w zbiorze danych treningowych, walidacyjnych i testowych

50% danych treningowych, 20% danych walidacyjnych, 30% danych testowych

```python
from sklearn.model_selection import train_test_split

X= df.drop(['salary'],axis=1)
y=df['salary']

split_size=0.3

#Tworzenie danych treningowych i testowych
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=split_size,random_state=22)

#Tworzenie danych treningowych i walidacyjnych
X_train, X_val, y_train, y_val = train_test_split(X_train,y_train,test_size=0.2,random_state=5)
```

```python
print("Dane treningowe: {0}{1}".format(X_train.shape, y_train.shape))
print("Dane walidacyjne: {0}{1}".format(X_val.shape, y_val.shape))
print("Dane testowe: {0}{1}".format(X_test.shape, y_test.shape))
```

```python
#Wybieranie kilku algorytmów do klasyfikacji
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
models = []
names = ['LR','Random Forest','Neural Network','GaussianNB','DecisionTreeClassifier','SVM',]

models.append((LogisticRegression()))
models.append((RandomForestClassifier(n_estimators=100)))
models.append((MLPClassifier()))
models.append((GaussianNB()))
models.append((DecisionTreeClassifier()))
models.append((SVC()))
```

```python
print(models)
```

```python
from sklearn import model_selection
from sklearn.metrics import accuracy_score
```

```python
kfold = model_selection.KFold(n_splits=5,random_state=7, shuffle=True)

for i in range(0,len(models)):    
    cv_result = model_selection.cross_val_score(models[i],X_train,y_train,cv=kfold,scoring='accuracy')
    score=models[i].fit(X_train,y_train)
    prediction = models[i].predict(X_val)
    acc_score = accuracy_score(y_val,prediction)     
    print ('-'*40)
    print ('{0}: {1}'.format(names[i],acc_score))
    
```

PRZECHODZIMY DALEJ Z ALGORYTMEM RANDOM FOREST, PONIEWAŻ WYKAZAŁ DOBRĄ DOKŁADNOŚĆ

PRZEWIDUJEMY NASZE DANE TESTOWE I ZOBACZMY WYNIK PROGNOZ

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
randomForest = RandomForestClassifier(n_estimators=100)
randomForest.fit(X_train,y_train)
prediction = randomForest.predict(X_test)
```

```python
print ('-'*40)
print ('Dokładność wyniku:')
print (accuracy_score(y_test,prediction))
print ('-'*40)
print ('Confusion Matrix:')
print (confusion_matrix(y_test,prediction))
print ('-'*40)
print ('Classification Matrix:')
print (classification_report(y_test,prediction))
```
